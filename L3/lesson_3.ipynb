2. Load the energy data from the file Energy Indicators.xls. It should be put into a DataFrame with the variable name of "energy". Exclude the footer and header information from the datafile

````python
energy = pd.read_excel("Energy Indicators.xls", header=None, skiprows=18, skipfooter=38)
energy
````

3. The first two columns are unneccessary, so you should get rid of them, and you should change the column labels so that the columns are:

    ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

````python
energy = energy.drop(energy.columns[[0, 1]], axis=1)
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy
````

4. Convert Energy Supply to gigajoules (there are 1,000,000 gigajoules in a petajoule). For all countries which have missing data (e.g. data with ...) make sure this is reflected as np.NaN values.

````python
energy["Energy Supply"] = pd.to_numeric(energy["Energy Supply"], errors="coerce")
energy["Energy Supply"] *= 1000000
pd.options.display.float_format = '{:,.2f}'.format # Display numbers in non-scientific format
energy
````

5. There are also several countries with numbers and/or parenthesis in their name. Be sure to remove these, e.g.:
```
    Bolivia (Plurinational State of) should be Bolivia,
    Switzerland17 should be Switzerland.
```

````python
energy['Original Country'] = energy['Country']
 
energy['Country'] = energy['Country'].apply(lambda x: re.sub(r'[^\w\s]', '', x))  # Delete all non-alphanumeric characters
energy['Country'] = energy['Country'].apply(lambda x: re.sub(r'\([^)]*\)', '', x))  # Delete everything inside the brackets
energy['Country'] = energy['Country'].apply(lambda x: re.sub(r'\d+', '', x))        # Delete numbers
energy['Country'] = energy['Country'].str.strip()  # Delete trailing spaces

changed_countries = energy[energy['Original Country'] != energy['Country']]

changed_countries[['Original Country', 'Country']]
````

10. Join the three datasets: Energy, GDP, and ScimEn into a new dataset (using the intersection of country names). Use only the 10 years (2006-2015) of GDP data and only the top 15 countries by Scimagojr 'Rank' (Rank 1 through 15).
The index of this DataFrame should be the name of the country, and the columns should be
['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', 2012', '2013', '2014', '2015']

````python
GDP = GDP[['Country Name'] + [str(year) for year in range(2006, 2016)]]
ScimEn = ScimEn[ScimEn['Rank'] <= 15]

# Merge the three datasets
merged_df = pd.merge(ScimEn, energy, how='inner', left_on='Country', right_on='Country')
merged_df = pd.merge(merged_df, GDP, how='inner', left_on='Country', right_on='Country Name')
merged_df = merged_df.set_index('Country')

merged_df.drop(columns=['Country Name'], inplace=True)

# Check the columns order
columns_order = ['Rank', 'Documents', 'Citable documents', 'Citations',
                 'Self-citations', 'Citations per document', 'H index', 
                 'Energy Supply', 'Energy Supply per Capita', '% Renewable'] + [str(year) for year in range(2006, 2016)]
merged_df = merged_df[columns_order]

merged_df
````

## Question 2

What is the average GDP over the last 10 years for each country? (exclude missing values from this calculation.)

This function should return a Series named avgGDP with 15 countries and their average GDP sorted in descending order.

````python
def answer_two():
    Top15 = merged_df
    
    # Mean of the GDP for the years 2006-2015
    avgGDP = Top15[[str(year) for year in range(2006, 2016)]].mean(axis=1, skipna=True)
    
    # Sort the values in descending order
    return avgGDP.sort_values(ascending=False)

answer_two()
````

## Question 3

By how much had the GDP changed over the 10 year span for the country with the 6th largest average GDP?

This function should return a single number.

````python
def answer_three():
    avgGDP = answer_two()
    
    # Get the 6th country
    country_6th = avgGDP.index[5]
    
    GDP_2006 = merged_df.loc[country_6th, '2006']
    GDP_2015 = merged_df.loc[country_6th, '2015']
    
    GDP_change = GDP_2015 - GDP_2006
    
    return float(GDP_change)

answer_three()
````

## Question 4

Create a new column that is the ratio of Self-Citations to Total Citations. What is the maximum value for this new column, and what country has the highest ratio?

This function should return a tuple with the name of the country and the ratio.


````python
def answer_four():
    merged_df['Self-Citations to Total Citations Ratio'] = merged_df['Self-citations'] / merged_df['Citations']
    
    max_ratio_country = merged_df['Self-Citations to Total Citations Ratio'].idxmax()
    max_ratio = merged_df.loc[max_ratio_country, 'Self-Citations to Total Citations Ratio']
    
    return (max_ratio_country, float(max_ratio))

answer_four()
````

## Question 5

Create a column that estimates the population using Energy Supply and Energy Supply per capita. What is the third most populous country according to this estimate?

This function should return a single string value.


````python
def answer_five():
    Top15 = merged_df.copy()
    
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    
    Top15 = Top15.sort_values(by='Population', ascending=False)
    
    third_most_populous = Top15.index[2]
    
    return third_most_populous

answer_five()
````

## Question 6

Create a column that estimates the number of citable documents per person. What is the correlation between the number of citable documents per capita and the energy supply per capita? Use the .corr() method, (Pearson's correlation).

This function should return a single number.


````python
def answer_six():
    Top15 = merged_df.copy()
    
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    
    Top15['Citable docs per capita'] = Top15['Citable documents'] / Top15['Population']
    
    correlation = Top15['Citable docs per capita'].corr(Top15['Energy Supply per Capita'])
    
    return float(correlation)

answer_six()
````

## Question 7

Use the following dictionary to group the Countries by Continent, then create a dateframe that displays the sample size (the number of countries in each continent bin), and the sum, mean, and std deviation for the estimated population of each country.

```
ContinentDict  = {'China':'Asia',
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
```
This function should return a DataFrame with index named Continent ['Asia', 'Australia', 'Europe', 'North America', 'South America'] and columns ['size', 'sum', 'mean', 'std']


````python
def answer_seven():
    Top15 = merged_df.copy()
    
    ContinentDict  = {'China':'Asia', 
                      'United States':'North America', 
                      'Japan':'Asia', 
                      'United Kingdom':'Europe', 
                      'Russian Federation':'Europe', 
                      'Canada':'North America', 
                      'Germany':'Europe', 
                      'India':'Asia',
                      'France':'Europe', 
                      'South Korea':'Asia', 
                      'Italy':'Europe', 
                      'Spain':'Europe', 
                      'Iran':'Asia',
                      'Australia':'Australia', 
                      'Brazil':'South America'}

    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']

    Top15['Continent'] = Top15.index.to_series().map(ContinentDict)

    # Group by continent and calculate the size, sum, mean and std
    result = Top15.groupby('Continent')['Population'].agg(['size', 'sum', 'mean', 'std'])
    
    return result

answer_seven()
````
